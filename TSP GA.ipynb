```python
#GA
import numpy as np
import random
from scipy.spatial.distance import squareform,pdist
import copy

"""評価関数"""
def evaluation(dist,order):
    return np.sum([dist[order[i],order[(i+1)%city_num]] for i in np.arange(city_num)])

"""選択"""
def select(evaluate,elite_gene_length):
    ind_list=list(np.argsort(eval_list))[::-1]#高い順にソート
    new_gene_list=[]
    for i in range(len(ini_gene)):
        new_gene_list.append(list(ini_gene[ind_list[i]]))
    select_gene_list=new_gene_list[:elite_gene_length]
    return select_gene_list

"""交叉（二点）"""
def crossover(gene1,gene2):
    #入れ替える位置
    idx=random.randint(0,city_num-1)
    jdx=random.randint(0,city_num-idx-1)
    child_gene_list=copy.deepcopy(gene1)#親固体をコピー
    child_gene_list[idx:jdx]=gene2[idx:jdx]
    return child_gene_list

"""突然変異"""
def mutate(gene,ind_mutate,gene_mutate,city_num):
    gene_list=[]
    for g in gene:
        if ind_mutate>(random.randint(0,100)/100):
            genom_list=[]
            for j in g:
                if gene_mutate>(random.randint(0,100)/100):
                    genom_list.append(random.randint(0,city_num))
                else:
                    genom_list.append(j)
            gene_list.append(g)
        else:
            gene_list.append(g)
            
    return gene_list

def main():
    """初期パラメータ"""
    city_num=5#個体数
    n_gene=15#一個体あたりの遺伝子の数
    elite_gene_length=5#エリート固体の数
    ind_mutate=0.1
    gene_mutate=0.1
    generation_num=30
    """初期固体生成"""
    ini_gene=np.zeros((n_gene,city_num),np.int)
    order=np.arange(city_num)
    for i in range(n_gene):
        np.random.shuffle(order)
        ini_gene[i]=order.copy()
    dist=squareform(pdist(np.random.random_sample((30,2))))
    ##fitnessによる評価loopにする
    for num in range(generation_num):
        print('--generation%i--'%(num+1))
        eval_list=[]
        for i in range(n_gene):
            eval_list.append(evaluation(dist,ini_gene[i]))
        select_gene_list=select(eval_list,elite_gene_length)
        child_list=[]
        for i in range(n_gene):
            c1=random.randint(0,city_num-1)
            c2=random.randint(0,city_num-1)
            child=list(crossover(select_gene_list[c1],select_gene_list[c2]))
            child_list.append(child)
            
        """次世代個体生成"""
        offspring=mutate(child_list[:n_gene-elite_gene_length],ind_mutate,gene_mutate,city_num)
        offspring.extend(select_gene_list)
        ini_gene=offspring
        print(ini_gene)
    
if __name__=='__main__':
    main()
    
            
                    
        
```
